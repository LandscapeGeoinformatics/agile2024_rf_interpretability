```python
import glob

import pandas as pd
```

```python
# Aggregation function list
agg_func_list = ["mean", "median"]
```

```python
agg_func = "mean"
files = glob.glob(f"experiments/agg_func_{agg_func}*.csv")
df = pd.concat([pd.read_csv(file) for file in files]).reset_index(drop=True)
display(df.sort_values("r2_test_reduced", ascending=False))
grouped = df.groupby(["test_size", "n_features_reduced"])["r2_test_reduced_gain"].median().reset_index()
top_test_size = grouped.sort_values("r2_test_reduced_gain", ascending=False).head(1)["test_size"].values[0]
top_n_features_reduced = grouped.sort_values("r2_test_reduced_gain", ascending=False).head(1)["n_features_reduced"].values[0]
df[(df["test_size"] == top_test_size) & (df["n_features_reduced"] == top_n_features_reduced)]
```
